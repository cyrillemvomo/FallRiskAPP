```python
#Closed eyes and foam data = fc
df_fc= pd.read_excel("/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/BDSinfo.xlsx")

df_fc= df_fc.loc[df_fc["Surface"].apply(lambda x: "oam" in x)]
df_fc= df_fc.loc[df_fc["Vision"].apply(lambda x: "losed" in x)]
# df_fc= df_fc.drop_duplicates(subset='Subject', keep="last") #TESTING IN PROGRESS
df_fc
df_fc['Trial_with_ext']=[suit + '.txt' for suit in df_fc.Trial.to_list()]
df_fc['FES_S'] = df_fc['FES_S'].str.replace('High Concern', '3')
df_fc['FES_S'] = df_fc['FES_S'].str.replace('Moderate Concern', '2')
df_fc['FES_S'] = df_fc['FES_S'].str.replace('Low Concern', '1')
df_fc['FES_S'] = df_fc['FES_S'].astype(float)
df_fc['IPAQ_S'] = df_fc['IPAQ_S'].str.replace('High', '3')
df_fc['IPAQ_S'] = df_fc['IPAQ_S'].str.replace('Moderate', '2')
df_fc['IPAQ_S'] = df_fc['IPAQ_S'].str.replace('Low', '1')
df_fc['IPAQ_S'] = df_fc['IPAQ_S'].astype(float)
df_fc['AgeGroup'] = df_fc['AgeGroup'].str.replace('Old', '1')
df_fc['AgeGroup'] = df_fc['AgeGroup'].str.replace('Young', '0')
df_fc['AgeGroup'] = df_fc['AgeGroup'].astype(float)
df_fc['Gender'] = df_fc['Gender'].str.replace('M', '1')
df_fc['Gender'] = df_fc['Gender'].str.replace('F', '0')
df_fc['Gender'] = df_fc['Gender'].astype(float)
df_fc

df_fc = df_fc.loc[:, ['Trial_with_ext', 'Subject', 'Vision', 'Surface', 'Age', 'AgeGroup', 'Gender','Height', 'Weight', 'BMI', 'Falls12m', 'Best_T', 'TMT_timeA', 'TMT_timeB', 'IPAQ_S', 'FES_T', 'FES_S']]
faller_fc = df_fc.loc[((df_fc['Best_T']<= 16)|(df_fc['FES_S']== 3)|(df_fc['IPAQ_S']== 1)|(df_fc['TMT_timeA']== 300)|(df_fc['TMT_timeB']== 300))&((df_fc['AgeGroup']== 1)&(df_fc['Falls12m']> 0))]# test and falls 
NonFaller_fc = df_fc.loc[((df_fc['Best_T']> 16)|(df_fc['FES_S']< 3)|(df_fc['IPAQ_S']> 1)|(df_fc['TMT_timeA']< 300)|(df_fc['TMT_timeB']< 300))&(df_fc['Falls12m']== 0)]# test and falls
NonFaller_fc['Label(0=Non-Faller|1=Faller)']=[0 for i in range(1,1+len(NonFaller_fc))]
faller_fc['Label(0=Non-Faller|1=Faller)']=[1 for i in range(1,1+len(faller_fc))]
df_fc = pd.concat([faller_fc, NonFaller_fc], axis=0)
df_fc.reset_index(drop=True, inplace=True)
# df_fc['ID']=[i for i in range(1,1+len(df_fc))]
df_fc['AgeGroup(0=young|1=old)'] = df_fc['AgeGroup'].to_list()
df_fc['Gender(0=F|1=M)'] = df_fc['Gender'].to_list()
df_fc = df_fc.loc[:, ['Trial_with_ext', 'Subject', 'Age', 'AgeGroup(0=young|1=old)', 'Gender(0=F|1=M)','Height', 'Weight', 'BMI', 'Falls12m', 'Best_T', 'FES_T', 'IPAQ_S', 'Label(0=Non-Faller|1=Faller)']]
df_fc = df_fc.drop_duplicates()
df_fc.to_csv('/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/MachineLEARNING/Info/Subjects_Info_fc.csv')
df_fc
a = df_fc.Subject.value_counts().to_list()
subj_list = df_fc.Subject.drop_duplicates().to_list()
print(subj_list, '/', a, '/', sum(a))


#copy file to specified folder
Fallers_Data_fc= df_fc[df_fc['Label(0=Non-Faller|1=Faller)'] == 1]
NonFallers_Data_fc= df_fc[df_fc['Label(0=Non-Faller|1=Faller)'] == 0]




import shutil, os
Fallers_Data_files_fc = Fallers_Data_fc.Trial_with_ext.to_list()

NonFallers_Data_files_fc = NonFallers_Data_fc.Trial_with_ext.to_list()



source = '/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/BDS'
dest1_fc = '/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/MachineLEARNING/Fallers_Data_fc'
dest2_fc = '/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/MachineLEARNING/Non_Fallers_Data_fc'

files = os.listdir(source)

def Movetree(src, dst, list_of_filename, symlinks=False, ignore=None):
    for item in list_of_filename:
        if item in os.listdir(src):
            s = os.path.join(src, item)
            d = os.path.join(dst, item)
            shutil.move(s, d)


Movetree(source, dest1_fc, Fallers_Data_files_fc)
Movetree(source, dest2_fc, NonFallers_Data_files_fc)
print(len(Fallers_Data_files_fc), len(NonFallers_Data_files_fc))
NonFallers_Data_fc.to_csv('/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/MachineLEARNING/Info/NonFallers_Info_fc.csv')
Fallers_Data_fc.to_csv('/Users/cyrilleetude/Desktop/Bureau - MacBook Air de Cyrille/Semester_Project_Cyrille_2022/MachineLEARNING/Info/Fallers_Info_fc.csv')
```
